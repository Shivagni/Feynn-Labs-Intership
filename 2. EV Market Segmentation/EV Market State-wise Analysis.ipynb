The dataset contains information about electric vehicles (EVs) across different states/regions in India. It has 25 rows and 8 columns, with the following details:

1. Region: Name of the State or region.
2. E-2 Wheelers: Count of electric two-wheelers.
3. E-3 Wheelers: Count of electric three-wheelers.
4. E-4 Wheelers: Count of electric four-wheelers.
5. E-Buses: Count of electric buses.
6. Total EVs: Total number of electric vehicles.
7. Total Charging Stations: Total number of charging stations.
8. Total Vehicles: Total number of vehicles (all types).

# **Exploratory Data Analysis**

```python
data.columns
```

```python
data['Percentage'] = data['Total EVs'] / data['Total Vehicles'] * 100
low3_ev = data.nsmallest(3, 'Percentage')
low3_ev
```

Since these States have lowest number of EVs wrt total vehicles in the State, we can infer that growth potential for EV markets in these states can be considered good.

```python
import altair as alt

# Melt the dataframe to long format
data_long = data.melt(id_vars=['Region'],
                      # These columns are the ones that will be "melted" or turned into rows
                     value_vars= ['E-2 Wheelers', 'E-3 Wheelers', 'E-4 Wheelers', 'E-Buses'],
                      #This new column will store the names of the melted columns (like 'E-2 Wheelers', 'E-3 Wheelers', etc.)
                     var_name='Vehicle Type',
                      # This new column will store the values from the melted columns.
                      value_name='Count')
# The result is a DataFrame where each row represents a specific vehicle type and its count for a given region.

# Create the bar chart
chart = alt.Chart(data_long).mark_bar().encode(
    x='Region',
    y='Count',
    color='Vehicle Type',
    tooltip=['Region', 'Vehicle Type', 'Count']
).properties(
    width=600,
    height=400
)
chart
```

```python
vehicle_counts = data[['E-2 Wheelers', 'E-3 Wheelers', 'E-4 Wheelers', 'E-Buses']].sum()

# Create the pie chart
plt.pie(vehicle_counts, labels=vehicle_counts.index, autopct='%1.1f%%', startangle=45)
plt.title('Distribution of Electric Vehicle Types')
plt.show()
```

2 wheelers are the most sold category of EVs

```python
top3_states = data.nlargest(3, 'Total EVs')

plt.pie(top3_states['Total EVs'], labels=top3_states['Region'], autopct='%1.1f%%', startangle=45)
plt.title('Top 3 States in Electric Vehicle Adoption')
plt.show()
```

```python
# 2 wheelers are the most sold category of EVs
top5_e2w = data.nlargest(5, 'E-2 Wheelers')

# Display top 5 Region in E-3 Wheelers
top5_e3w = data.nlargest(5, 'E-3 Wheelers')

# Display top 5 Region in E-4 Wheelers
top5_e4w = data.nlargest(5, 'E-4 Wheelers')

# Display top 5 Region in E-Buses
top5_ebuses = data.nlargest(5, 'E-Buses')

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(6,6))

# Plot the pie charts
axes[0, 0].pie(top5_e2w['E-2 Wheelers'], labels=top5_e2w['Region'], autopct='%1.1f%%')
axes[0, 0].set_title('Top 5 Regions for E-2 Wheelers')

axes[0, 1].pie(top5_e3w['E-3 Wheelers'], labels=top5_e3w['Region'], autopct='%1.1f%%')
axes[0, 1].set_title('Top 5 Regions for E-3 Wheelers')

axes[1, 0].pie(top5_e4w['E-4 Wheelers'], labels=top5_e4w['Region'], autopct='%1.1f%%')
axes[1, 0].set_title('Top 5 Regions for E-4 Wheelers')

axes[1, 1].pie(top5_ebuses['E-Buses'], labels=top5_ebuses['Region'], autopct='%1.1f%%')
axes[1, 1].set_title('Top 5 Regions for E-Buses')


# Adjust layout and display
plt.tight_layout()

plt.show()

```

States Leading in :
*   2 Wheelers - Maharashtra, Karnataka, Tamil Nadu
*   3 Wheelers - Uttar Pradesh, Bihar, Assam
*   4 Wheelers - Maharashtra, Telangana, Karnataka
*   E-Buses    - Maharashtra, Karnataka, Tamil Nadu
