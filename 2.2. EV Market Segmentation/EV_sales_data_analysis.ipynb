```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## **EV Sales data Analysis**

```python
data = pd.read_csv("/content/sample_data/EV Market Data.csv")
data
```

```python
data1 = data.iloc[:, :-2]
data1.drop([4],inplace=True)
data1
```

```python
# Transpose the DataFrame for easier plotting
df_transposed = data1.set_index('Category').transpose()

# Create the line chart
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for category in df_transposed.columns:
    plt.plot(df_transposed.index, df_transposed[category], label=category)
    # Add numbers on the chart
    for x, y in zip(df_transposed.index, df_transposed[category]):
        plt.text(x, y, str(y), ha='center', va='bottom')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Trend by Category')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data_24_25 = data[['Category','24-25 (till 12 Aug)']]
data_24_25.drop([4],inplace=True)
data_24_25
```

```python
# Extracting data from the DataFrame
categories = data_24_25['Category']
values = data_24_25['24-25 (till 12 Aug)']

# Creating the bar chart
plt.bar(categories, values)
plt.xlabel('Category')
plt.ylabel('Sales till 12 Aug 2024')

# Adding text on each bar
for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Displaying the chart
plt.show()
```
